```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
```

```python
suhu_udara = ctrl.Antecedent(np.arange(0, 31, 1), "suhu_udara")
kelembaban_udara = ctrl.Antecedent(np.arange(0, 91, 1), "kelembaban_udara")
kecepatan_angin = ctrl.Antecedent(np.arange(0, 7, 1), "kecepatan_angin")

# Variable input
suhu_udara['dingin'] = fuzz.trimf(suhu_udara.universe, [0, 0, 15])
suhu_udara['normal'] = fuzz.trimf(suhu_udara.universe, [0, 15, 30])
suhu_udara['panas'] = fuzz.trimf(suhu_udara.universe, [15, 30, 30])

kelembaban_udara['kering'] = fuzz.trimf(kelembaban_udara.universe, [0, 0, 45])
kelembaban_udara['lembab'] = fuzz.trimf(kelembaban_udara.universe, [0, 45, 90])
kelembaban_udara['basah'] = fuzz.trimf(kelembaban_udara.universe, [45, 90, 90])

kecepatan_angin['pelan'] = fuzz.trimf(kecepatan_angin.universe, [0, 0, 3])
kecepatan_angin['sedang'] = fuzz.trimf(kecepatan_angin.universe, [0, 3, 6])
kecepatan_angin['kencang'] = fuzz.trimf(kecepatan_angin.universe, [3, 6, 6])

suhu_udara.view()
kelembaban_udara.view()
kecepatan_angin.view()
```

# 

### Berdasarkan data 1 Januari 2023

```python
curah_hujan = ctrl.Consequent(np.arange(0, 111, 1), "curah_hujan")

# Variable output
curah_hujan['rendah'] = fuzz.trimf(curah_hujan.universe, [20, 20, 50])
curah_hujan['sedang'] = fuzz.trimf(curah_hujan.universe, [55, 65, 80])
curah_hujan['tinggi'] = fuzz.trimf(curah_hujan.universe, [81, 110, 110])

# Fuzzy rules
rule1 = ctrl.Rule(suhu_udara['dingin'] & kecepatan_angin['pelan'] & kelembaban_udara['kering'], curah_hujan['rendah'])
rule2 = ctrl.Rule(suhu_udara['dingin'] & kecepatan_angin['pelan'] & kelembaban_udara['lembab'], curah_hujan['sedang'])
rule3 = ctrl.Rule(suhu_udara['dingin'] & kecepatan_angin['pelan'] & kelembaban_udara['basah'], curah_hujan['tinggi'])
rule4 = ctrl.Rule(suhu_udara['dingin'] & kecepatan_angin['sedang'] & kelembaban_udara['kering'], curah_hujan['rendah'])
rule5 = ctrl.Rule(suhu_udara['dingin'] & kecepatan_angin['sedang'] & kelembaban_udara['lembab'], curah_hujan['tinggi'])
rule6 = ctrl.Rule(suhu_udara['dingin'] & kecepatan_angin['sedang'] & kelembaban_udara['basah'], curah_hujan['tinggi'])
rule7 = ctrl.Rule(suhu_udara['dingin'] & kecepatan_angin['kencang'] & kelembaban_udara['kering'], curah_hujan['sedang'])
rule8 = ctrl.Rule(suhu_udara['dingin'] & kecepatan_angin['kencang'] & kelembaban_udara['lembab'], curah_hujan['tinggi'])
rule9 = ctrl.Rule(suhu_udara['dingin'] & kecepatan_angin['kencang'] & kelembaban_udara['basah'], curah_hujan['tinggi'])
rule10 = ctrl.Rule(suhu_udara['normal'] & kecepatan_angin['pelan'] & kelembaban_udara['kering'], curah_hujan['rendah'])
rule11 = ctrl.Rule(suhu_udara['normal'] & kecepatan_angin['pelan'] & kelembaban_udara['lembab'], curah_hujan['rendah'])
rule12 = ctrl.Rule(suhu_udara['normal'] & kecepatan_angin['pelan'] & kelembaban_udara['basah'], curah_hujan['sedang'])
rule13 = ctrl.Rule(suhu_udara['normal'] & kecepatan_angin['sedang'] & kelembaban_udara['kering'], curah_hujan['rendah'])
rule14 = ctrl.Rule(suhu_udara['normal'] & kecepatan_angin['sedang'] & kelembaban_udara['lembab'], curah_hujan['sedang'])
rule15 = ctrl.Rule(suhu_udara['normal'] & kecepatan_angin['sedang'] & kelembaban_udara['basah'], curah_hujan['tinggi'])
rule16 = ctrl.Rule(suhu_udara['normal'] & kecepatan_angin['kencang'] & kelembaban_udara['kering'], curah_hujan['sedang'])
rule17 = ctrl.Rule(suhu_udara['normal'] & kecepatan_angin['kencang'] & kelembaban_udara['lembab'], curah_hujan['tinggi'])
rule18 = ctrl.Rule(suhu_udara['normal'] & kecepatan_angin['kencang'] & kelembaban_udara['basah'], curah_hujan['tinggi'])
rule19 = ctrl.Rule(suhu_udara['panas'] & kecepatan_angin['pelan'] & kelembaban_udara['kering'], curah_hujan['rendah'])
rule20 = ctrl.Rule(suhu_udara['panas'] & kecepatan_angin['pelan'] & kelembaban_udara['lembab'], curah_hujan['rendah'])
rule21 = ctrl.Rule(suhu_udara['panas'] & kecepatan_angin['pelan'] & kelembaban_udara['basah'], curah_hujan['sedang'])
rule22 = ctrl.Rule(suhu_udara['panas'] & kecepatan_angin['sedang'] & kelembaban_udara['kering'], curah_hujan['rendah'])
rule23 = ctrl.Rule(suhu_udara['panas'] & kecepatan_angin['sedang'] & kelembaban_udara['lembab'], curah_hujan['sedang'])
rule24 = ctrl.Rule(suhu_udara['panas'] & kecepatan_angin['sedang'] & kelembaban_udara['basah'], curah_hujan['sedang'])
rule25 = ctrl.Rule(suhu_udara['panas'] & kecepatan_angin['kencang'] & kelembaban_udara['kering'], curah_hujan['rendah'])
rule26 = ctrl.Rule(suhu_udara['panas'] & kecepatan_angin['kencang'] & kelembaban_udara['lembab'], curah_hujan['sedang'])
rule27 = ctrl.Rule(suhu_udara['panas'] & kecepatan_angin['kencang'] & kelembaban_udara['basah'], curah_hujan['tinggi'])

# Sistem inferensi fuzzy
prediksi_curah_hujan_ctrl = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5, rule6, rule7, rule8, rule9, rule10,
                                               rule11, rule12, rule13, rule14, rule15, rule16, rule17, rule18, rule19, 
                                               rule20, rule21, rule22, rule23, rule24, rule25, rule26, rule27])
prediksi_curah_hujan = ctrl.ControlSystemSimulation(prediksi_curah_hujan_ctrl)

# Input variable data pada 1 Januari 2023
prediksi_curah_hujan.input['suhu_udara'] = 27.5
prediksi_curah_hujan.input['kelembaban_udara'] = 79
prediksi_curah_hujan.input['kecepatan_angin'] = 4

prediksi_curah_hujan.compute()

print("Prediksi Curah Hujan = ", prediksi_curah_hujan.output['curah_hujan'])

curah_hujan.view(sim = prediksi_curah_hujan)
```
