```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import math
```

```python

# Fungsi untuk menghitung turunan y' = e * y
def f(y):
    return 0.5 * y

# Inisialisasi variabel
t0 = 0  # Waktu awal
t_akhir = 2  # Waktu akhir
N = 20  # Jumlah interval
dt = 0.1  # Langkah waktu
y_awal = 10  # Nilai awal y


# List untuk menyimpan nilai y dan t
y_values = [y_awal]
t_values = [t0]

# Iterasi menggunakan metode Euler
for n in range(N):
    y_sekarang = y_values[-1]  # Nilai y terakhir (sekarang)
    y_baru = y_sekarang + dt * f(y_sekarang)  # Rumus metode Euler
    t_baru = t_values[-1] + dt  # Perbarui nilai t
    
    # Simpan nilai baru ke dalam list
    y_values.append(y_baru)
    t_values.append(t_baru)

    
iterasi = np.arange(len(y_values))  # Membuat array nomor iterasi mulai dari 0 hingga jumlah elemen x_nilai
    
    # Membuat dictionary yang akan digunakan untuk membentuk DataFrame
data = {
    'Iterasi': '',  # Kolom untuk nomor iterasi
    'x': y_values,  # Kolom untuk nilai x pada setiap iterasi
}
    
df = pd.DataFrame(data)  # Membuat DataFrame dari dictionary data
print(df)  # Menampilkan DataFrame sebagai tabel hasil iterasi

# Solusi eksak y(t) = 10 * e^(0.5 * t)
y_exact = [10 * math.exp(0.5 * t) for t in t_values]

# Plot aproksimasi Euler
plt.plot(t_values, y_values, label='Aproksimasi Euler', color='blue')

# Plot solusi eksak
plt.plot(t_values, y_exact, label='Solusi Eksak', linestyle='--', color='green')

# Scatter plot untuk menampilkan titik-titik aproksimasi Euler
plt.scatter(t_values, y_values, color='red')

plt.title("Plot nilai t dan y")
plt.ylabel("f(t, y)")
plt.xlabel("t")
plt.show()

```
